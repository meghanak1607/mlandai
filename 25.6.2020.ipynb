```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading Boston dataset
from sklearn.datasets import load_boston
boston = load_boston()
```

```python
boston_df = pd.DataFrame(boston.data,columns=boston.feature_names)
boston_df.head()
```

```python
boston_df['MEDV']= boston.target   #adding output column to dataset
boston_df.head()
```

```python
#  to find the corelation we use a function called corr()
boston_df.corr()
```

```python
# visualizing the corelation
sns.heatmap(boston_df.corr())
```

```python
plt.figure(figsize=(14,14))                          #to enlarge the figure and make it clear
sns.heatmap(boston_df.corr(),annot=True,cmap='YlGnBu')    #to get values and colours
```

```python
# to get correlation > 0.7
correlation = boston_df.corr()

cor = correlation[correlation > 0.7]
plt.figure(figsize=(14,14))
sns.heatmap(cor,cmap='Greens',annot=True)
```

```python
#checking the correlation between 0.7 and 0.9
cor = correlation[(correlation>0.7)&(correlation<0.9)]
cor
```

# Building the model

```python
#split the data into training and testing, build the model on training data and test the model on testing dataset
```

```python
## Split the data into input and output
X = boston_df.drop(['MEDV'],axis=1)  #input columns , removing the output and separately storing
y= boston_df.MEDV 
```

```python
# Divide the X and y into train and test
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25, random_state = 0)
```

```python
# Build the model on Training data--> X_train and y_train
# Sklearn library : import , instantiate,fit
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X_train,y_train)
```

```python
# Intercept and the coefficient values
print(lm.intercept_)
lm.coef_
```

```python
# Create a dataframe for coefficients(as it is in array format)
coefficients = pd.DataFrame([X_train.columns,lm.coef_]).T
coefficients
```

```python
# checking the model prediction on training data
y_train_pred = lm.predict(X_train)
y_train_pred
```

```python
# Test the model on testing data
y_test_pred = lm.predict(X_test)  # testdata --> unseen data
y_test_pred
```
