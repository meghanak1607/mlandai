### LINEAR ALGEBRA

```python
from scipy import linalg
#create 2d matrix
ar2 = np.array([[4,5],[3,2]])
linalg.det(ar2)
```

#### IMAGE OPERATIONS

```python
from scipy import misc
data= misc.face()    #to load the picture
plt.imshow(data)
```

```python
data.shape,type(data)
```

```python
    #for flipping image
flip_data = np.flipud(data)
plt.imshow(flip_data)
```

```python
rand_arr = np.random.randint(low=0,high=255,size=data.shape)
plt.imshow(rand_arr)
```

```python
     #for rotating image
from scipy import ndimage    #import subpackage
data_rotate = ndimage.rotate(data,45)
plt.imshow(data_rotate)
```
