```python
##To create a dataframe out of dictionary
boston_df = pd.DataFrame(data = boston.data,columns= boston.feature_names)
boston_df
```

```python
# created a new column and passed target values(output column)  into the dictionary
boston_df['MEDV']=boston.target 
boston_df.head()
```

```python
#out of 13 columns, we will work with single input(RM-->avg no. of rooms) and medv as output
# preparing the input and output columns
X=boston_df[['RM']] #selecting column as dataframe
y=boston_df['MEDV'] #output will always be a single column ,so series
```

```python
boston_df['MEDV'] #selecting column as series
```

```python
#plt.style.available
with plt.style.context("grayscale"):
    plt.plot(X,y,'bo')  # bo---> blue color , circle as marker
    plt.xlabel("RM I/P")
    plt.ylabel("MEDV O/P")
    plt.title("Scatter Plot")
```

```python
# usual sklearn: import , instantiate,fit
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X,y)  #objectname.fit(input,output)
```

```python
# print the intercept and coefficient
print(lm.intercept_)
print(lm.coef_)
```

```python
# y=mx+c
#m value --> coefficient(9.10210898)
#c value--> intercept(-34.67062077643857)
(9.10210898 * 6.575)+(-34.67062077643857)
```

# interpreting the model:
a unit increase in X(RM) is associated with a 9.10210898 unit increase in MEDV

```python
## Pipeline--> first we create a model on the  data
## second , we need to compare the model(predicted values with actual values)
```

```python
# with thw help of predict, we are going to find our predicted values
y_pred = lm.predict(X)
y_pred
```

```python
##r2_value-->to check the model performance
from sklearn.metrics import r2_score
r2_score(y,y_pred)
```

```python
#Observe the regression line
with sns.axes_style("dark"):#temporary styling 
    sns.regplot(X,y)
```

```python
#scatterplot(Regression)
#first we are making a scatter plot of the original data(X and Y)
#we need to plot a line for the predicted values(X,y_pred)
plt.scatter(X,y,color='red')
plt.plot(X,y_pred,color='black')
plt.xlabel("RM I/P")
plt.ylabel("MEDV O/P")
```
